```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors 
from scipy import stats


%config InlineBackend.figure_format = 'retina'
```

```python
df_cars = pd.read_csv("cars.csv")
df_cars.head()
```

```python
red = '#ff0000'
black = '#000000'

fig, ax = plt.subplots(figsize=(6,4))

#set color to be mapped colors
ours = df_cars[(df_cars['MPG']>=40) & (df_cars['WGT']<=2000)]
others = df_cars[(df_cars['MPG']<40) & (df_cars['WGT']>2000)]
ax.scatter(ours['WGT'], ours['MPG'], label="ours", color=red, s=40, alpha=.4)
ax.scatter(others['WGT'], others['MPG'], label="others", color=black, s=40, alpha=.4)

#Calculate the mean MPG and mean WGT
my = df_cars['MPG']
mx = df_cars['WGT']


slope, intercept, r_value, p_value, std_err = stats.linregress(mx, my)
a = intercept # 46.21652431150987
b = slope #-0.007647342458584569 
y = (mx*b + a)
plt.plot(mx, y, 'g', label='MPG=a+b*Weight')


# Show legend for the label of the line
ax.legend(loc="upper right")

#Add title, xlabel and ylabel
ax.set_xlabel("Weight(lb)",fontsize=13, fontname="Times")
ax.set_ylabel("Miles per gallon",fontsize=13, fontname="Times")
ax.set_title("Scatterplot of Weight vs MPG",c='grey', weight = 'bold')


#set x label position to bottom right
ax.xaxis.set_label_coords(0.9, -0.1) 
#set y label position to top left
ax.yaxis.set_label_coords(-0.1, 0.80)

#hide the frames, only show x and y axis
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```
